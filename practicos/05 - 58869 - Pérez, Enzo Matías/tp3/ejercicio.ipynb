## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def unidades_vendidas(datos):
  
    ventas_por_producto = {}
    
    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']
        
        
        if producto in ventas_por_producto:
            ventas_por_producto[producto] += cantidad
        else:
          
            ventas_por_producto[producto] = cantidad
    
    return ventas_por_producto

def listar_ventas(ventas):
   
    for producto, cantidad in ventas.items():
        print(f"{producto}: {cantidad} unidades")


ventas = unidades_vendidas(datos)
listar_ventas(ventas)
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def precio_promedio(datos):
    precios_por_producto = {}
    

    for venta in datos:
        producto = venta['producto']
        precio = venta['precio']
        
        
        if producto in precios_por_producto:
            precios_por_producto[producto]['total_precio'] += precio
            precios_por_producto[producto]['conteo'] += 1
        else:
          
            precios_por_producto[producto] = {'total_precio': precio, 'conteo': 1}
    
   
    promedios = {producto: data['total_precio'] / data['conteo'] 
                 for producto, data in precios_por_producto.items()}
    
    return promedios

def listar_precios(promedios):

    for producto, promedio in promedios.items():
        print(f"{producto}: ${promedio:.2f} de precio promedio")


precios = precio_promedio(datos)
listar_precios(precios)

````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def ranking_productos(datos, top=3):
   
    ventas_por_producto = {}
    
    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']
        
        
        if producto in ventas_por_producto:
            ventas_por_producto[producto] += cantidad
        else:
            ventas_por_producto[producto] = cantidad
    
    
    ranking = sorted(ventas_por_producto.items(), key=lambda item: item[1], reverse=True)
    
   
    return ranking[:top]

def listar_ranking(ranking):
    
    for i, (producto, cantidad) in enumerate(ranking, start=1):
        print(f"{i}. {producto}: {cantidad} unidades")

                     
ranking = ranking_productos(datos)
listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
def ventas_por_mes(datos):
   
    ventas_mensuales = {}
    
    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']
        fecha = venta['fecha']
        
      
        mes = fecha[:7]
        
      
        if mes not in ventas_mensuales:
            ventas_mensuales[mes] = {}
        
       
        if producto in ventas_mensuales[mes]:
            ventas_mensuales[mes][producto] += cantidad
        else:
            ventas_mensuales[mes][producto] = cantidad
    
    return ventas_mensuales

def listar_ventas_mensuales(ventas):
    
    for mes, productos in ventas.items():
        print(f"Ventas de {mes}:")
        for producto, cantidad in productos.items():
            print(f"  {producto}: {cantidad} unidades")
        print()  

ventas = ventas_por_mes(datos)
listar_ventas_mensuales(ventas)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def resumen_ventas(datos):

    resumen = {}
    
    for venta in datos:
        producto = venta['producto']
        precio = venta['precio']
        cantidad = venta['cantidad']
        importe = precio * cantidad
        
      
        if producto in resumen:
            resumen[producto]['total_cantidad'] += cantidad
            resumen[producto]['total_importe'] += importe
            resumen[producto]['total_precio'] += precio
            resumen[producto]['conteo'] += 1
        else:
          
            resumen[producto] = {
                'total_cantidad': cantidad,
                'total_importe': importe,
                'total_precio': precio,
                'conteo': 1
            }
    
  
    for producto, datos in resumen.items():
        datos['precio_promedio'] = datos['total_precio'] / datos['conteo']
    
    return resumen

def informe_ventas(resumen):

    productos_ordenados = sorted(resumen.items())
    
    print("Informe de ventas:")
    print(f"{'Producto':<15} {'Unidades':<10} {'Importe Total':<15} {'Precio Promedio':<15}")
    print("-" * 60)
    

    for producto, datos in productos_ordenados:
        total_cantidad = datos['total_cantidad']
        total_importe = datos['total_importe']
        precio_promedio = datos['precio_promedio']
        
        print(f"{producto:<15} {total_cantidad:<10} ${total_importe:<15.2f} ${precio_promedio:<15.2f}")

resumen = resumen_ventas(datos)
informe_ventas(resumen)

````
