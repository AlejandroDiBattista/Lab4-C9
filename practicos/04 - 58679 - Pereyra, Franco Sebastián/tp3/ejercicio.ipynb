## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def obtener_informacion():

    filas = open("datos.dat", "r").readlines()

    registros = []

    for fila in filas:

        dia = fila[0:10].strip()
        articulo = fila[10:40].strip()
        costo = int(fila[40:50].strip())
        numero = int(fila[50:55].strip()) 

        entrada = {
            "dia": dia,
            "articulo": articulo,
            "costo": costo,
            "numero": numero
        }

        registros.append(entrada)

    return registros


informacion = obtener_informacion()


for entrada in informacion:
    print(entrada)
````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def sumar_totales(registros):
    suma_importe = 0
    suma_cantidad = 0

    for entrada in registros:

        total = entrada['costo'] * entrada['numero']
        suma_importe += total
        suma_cantidad += entrada['numero']

    return suma_importe, suma_cantidad


importe_total, cantidad_total = sumar_totales(informacion)

print(f"Las ventas fueron de ${importe_total:.2f} en {cantidad_total} unidades.")
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def calcular_precio_promedio(registros):

    total_precios = {}
    total_cantidades = {}

    for entrada in registros:
        articulo = entrada['articulo']
        costo = entrada['costo']
        numero = entrada['numero']

        if articulo in total_precios:
            total_precios[articulo] += costo * numero
            total_cantidades[articulo] += numero
        else:
            total_precios[articulo] = costo * numero
            total_cantidades[articulo] = numero

    promedio_precios = {}
    for articulo in total_precios:
        promedio_precios[articulo] = total_precios[articulo] / total_cantidades[articulo]

    return promedio_precios

def mostrar_precios(promedios):

    print("Precio promedio por producto:")
    for articulo, costo in promedios.items():
        print(f"{articulo}: ${costo:.2f} por unidad")


precios_promedio = calcular_precio_promedio(informacion)
mostrar_precios(precios_promedio)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def clasificar_productos(registros, limite=3):

    ventas_por_articulo = {}

    for entrada in registros:
        articulo = entrada['articulo']
        numero = entrada['numero']

        if articulo in ventas_por_articulo:
            ventas_por_articulo[articulo] += numero
        else:
            ventas_por_articulo[articulo] = numero

    clasificacion = sorted(ventas_por_articulo.items(), key=lambda x: x[1], reverse=True)

    return clasificacion[:limite]

def mostrar_clasificacion(clasificacion):

    print("Ranking de productos m√°s vendidos:")
    for articulo, numero in clasificacion:
        print(f"{articulo}: {numero} unidades vendidas")


clasificacion_productos = clasificar_productos(informacion)
mostrar_clasificacion(clasificacion_productos)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
def calcular_ventas_mensuales(registros):

    ventas_por_periodo = {}

    for entrada in registros:
        dia = entrada['dia']
        articulo = entrada['articulo']
        numero = entrada['numero']

        anio_mes = dia[:7]

        if anio_mes not in ventas_por_periodo:
            ventas_por_periodo[anio_mes] = {}

        if articulo in ventas_por_periodo[anio_mes]:
            ventas_por_periodo[anio_mes][articulo] += numero
        else:
            ventas_por_periodo[anio_mes][articulo] = numero

    return ventas_por_periodo

def mostrar_ventas_mensuales(ventas):

    print("Ventas por mes:")
    for periodo, articulos in ventas.items():
        print(f"\nMes: {periodo}")
        for articulo, numero in articulos.items():
            print(f"{articulo}: {numero} unidades vendidas")


ventas_mensuales = calcular_ventas_mensuales(informacion)
mostrar_ventas_mensuales(ventas_mensuales)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def generar_resumen_ventas(registros):

    informe = {}

    for entrada in registros:
        articulo = entrada['articulo']
        costo = entrada['costo']
        numero = entrada['numero']
        total = costo * numero

        if articulo in informe:
            informe[articulo]['numero'] += numero
            informe[articulo]['total_importe'] += total
        else:
            informe[articulo] = {
                'costo_promedio': costo,
                'numero': numero,
                'total_importe': total
            }

    for articulo in informe:
        informe[articulo]['costo_promedio'] = informe[articulo]['total_importe'] / informe[articulo]['numero']

    return informe

def mostrar_informe_ventas(informe):

    articulos_ordenados = sorted(informe.items(), key=lambda x: x[0])

    print("Informe General de Ventas:")
    print(f"{'Producto':<20} {'Costo Promedio':<20} {'Unidades Vendidas':<20} {'Importe Total':<15}")
    print("-" * 80)
    for articulo, datos in articulos_ordenados:
        print(f"{articulo:<20} ${datos['costo_promedio']:<20.2f} {datos['numero']:<20} ${datos['total_importe']:<14.2f}")


resumen_ventas = generar_resumen_ventas(informacion)
mostrar_informe_ventas(resumen_ventas)
````
