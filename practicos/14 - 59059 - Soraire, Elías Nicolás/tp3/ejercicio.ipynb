## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def cargar_datos():
    lineas = open("datos.dat", "r").readlines() 
    datos = []
    for linea in lineas:
        fecha = linea[0:10].strip()
        producto = linea[10:40].strip()
        precio = float(linea[40:50].strip())
        cantidad = int(linea[50:55].strip())
        datos.append({"fecha": fecha, "producto": producto, "precio": precio, "cantidad": cantidad})
    return datos

def calcular_totales(datos):
    total_importe = 0
    total_cantidad = 0

    for registro in datos:
        total_importe += registro['precio'] * registro['cantidad']
        total_cantidad += registro['cantidad']

    return total_importe, total_cantidad

datos = cargar_datos()
importe, cantidad = calcular_totales(datos)


print(f"Las ventas fueron de ${importe:.2f} en {cantidad} unidades")
````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def cargar_datos():
    lineas = open("datos.dat", "r").readlines()
    datos = []
    for linea in lineas:
        registro = {
            'fecha': linea[:10].strip(),
            'producto': linea[10:40].strip(),
            'precio': float(linea[40:50].strip()),
            'cantidad': int(linea[50:55].strip())
        }
        datos.append(registro)
    return datos

def unidades_vendidas(datos):
    ventas_por_producto = {}
    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']
        if producto in ventas_por_producto:
            ventas_por_producto[producto] += cantidad
        else:
            ventas_por_producto[producto] = cantidad
    return ventas_por_producto

def listar_ventas(ventas):
    for producto, cantidad in ventas.items():
        print(f"Producto: {producto} - Unidades Vendidas: {cantidad}")


datos = cargar_datos()

ventas = unidades_vendidas(datos)

listar_ventas(ventas)
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def cargar_datos():
    lineas = open("datos.dat", "r").readlines()
    datos = []
    for linea in lineas:
        registro = {
            'fecha': linea[:10].strip(),
            'producto': linea[10:40].strip(),
            'precio': float(linea[40:50].strip()),
            'cantidad': int(linea[50:55].strip())
        }
        datos.append(registro)
    return datos

def precio_promedio(datos):
    total_por_producto = {}
    conteo_por_producto = {}

    for venta in datos:
        producto = venta['producto']
        precio = venta['precio']
        
        if producto in total_por_producto:
            total_por_producto[producto] += precio
            conteo_por_producto[producto] += 1
        else:
            total_por_producto[producto] = precio
            conteo_por_producto[producto] = 1

    promedio_por_producto = {}
    for producto in total_por_producto:
        promedio_por_producto[producto] = total_por_producto[producto] / conteo_por_producto[producto]

    return promedio_por_producto

def listar_precios(precios):
    for producto, promedio in precios.items():
        print(f"Producto: {producto} - Precio Promedio: ${promedio:.2f}")

datos = cargar_datos()

precios = precio_promedio(datos)

listar_precios(precios)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def cargar_datos():
    lineas = open("datos.dat", "r").readlines()
    datos = []
    for linea in lineas:
        registro = {
            'fecha': linea[:10].strip(),
            'producto': linea[10:40].strip(),
            'precio': float(linea[40:50].strip()),
            'cantidad': int(linea[50:55].strip())
        }
        datos.append(registro)
    return datos

def ranking_productos(datos, top=3):
    unidades_por_producto = {}
    
    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']
        
        if producto in unidades_por_producto:
            unidades_por_producto[producto] += cantidad
        else:
            unidades_por_producto[producto] = cantidad

    ranking = sorted(unidades_por_producto.items(), key=lambda x: x[1], reverse=True)
    return ranking[:top]

def listar_ranking(ranking):
    print("Ranking de Productos MÃ¡s Vendidos:")
    for producto, cantidad in ranking:
        print(f"Producto: {producto} - Unidades Vendidas: {cantidad}")


datos = cargar_datos()

ranking = ranking_productos(datos)

listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
from collections import defaultdict
from datetime import datetime

def cargar_datos():
    lineas = open("datos.dat", "r").readlines()
    datos = []
    for linea in lineas:
        registro = {
            'fecha': linea[:10].strip(),
            'producto': linea[10:40].strip(),
            'precio': float(linea[40:50].strip()),
            'cantidad': int(linea[50:55].strip())
        }
        datos.append(registro)
    return datos

def ventas_por_mes(datos):
    ventas_mensuales = defaultdict(lambda: defaultdict(int))

    for venta in datos:
        fecha = venta['fecha']
        producto = venta['producto']
        cantidad = venta['cantidad']

        mes = datetime.strptime(fecha, "%Y-%m-%d").strftime("%Y-%m")
        
        ventas_mensuales[mes][producto] += cantidad

    return dict(ventas_mensuales)

def listar_ventas_mensuales(ventas):
    for mes, productos in ventas.items():
        print(f"\nVentas en {mes}:")
        for producto, cantidad in productos.items():
            print(f"Producto: {producto} - Unidades Vendidas: {cantidad}")


datos = cargar_datos()

ventas = ventas_por_mes(datos)

listar_ventas_mensuales(ventas)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
from collections import defaultdict

def cargar_datos():
    lineas = open("datos.dat", "r").readlines()
    datos = []
    for linea in lineas:
        registro = {
            'fecha': linea[:10].strip(),
            'producto': linea[10:40].strip(),
            'precio': float(linea[40:50].strip()),
            'cantidad': int(linea[50:55].strip())
        }
        datos.append(registro)
    return datos

def resumen_ventas(datos):
    ventas_resumen = defaultdict(lambda: {'total_precio': 0, 'total_cantidad': 0, 'total_importe': 0})
    
    for venta in datos:
        producto = venta['producto']
        precio = venta['precio']
        cantidad = venta['cantidad']
        importe = precio * cantidad
        
        ventas_resumen[producto]['total_precio'] += precio * cantidad
        ventas_resumen[producto]['total_cantidad'] += cantidad
        ventas_resumen[producto]['total_importe'] += importe
    
    for producto, datos in ventas_resumen.items():
        if datos['total_cantidad'] > 0:
            datos['precio_promedio'] = datos['total_precio'] / datos['total_cantidad']
        else:
            datos['precio_promedio'] = 0
    
    return dict(ventas_resumen)

def informe_ventas(resumen):
    productos_ordenados = sorted(resumen.keys())
    
    print(f"{'Producto':<30} {'Precio Promedio':<15} {'Cantidad Vendida':<18} {'Total Vendido':<15}")
    print("-" * 80)
    
    for producto in productos_ordenados:
        datos = resumen[producto]
        print(f"{producto:<30} {datos['precio_promedio']:<15.2f} {datos['total_cantidad']:<18} {datos['total_importe']:<15.2f}")


datos = cargar_datos()

resumen = resumen_ventas(datos)

informe_ventas(resumen)
````
