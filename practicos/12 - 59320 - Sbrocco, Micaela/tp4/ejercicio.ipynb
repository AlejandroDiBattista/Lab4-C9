# SimulaciÃ³n del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquÃ­a de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    valores_especiales = {
        (1, "espada"): 14,
        (1, "basto"): 13,
        (7, "espada"): 12,
        (7, "oro"): 11,
    }

    if (numero, palo) in valores_especiales:
        return valores_especiales[(numero, palo)]
    
    if numero == 3:
        return 10
    elif numero == 2:
        return 9
    elif numero == 1:
        return 8
    elif numero == 12:
        return 7
    elif numero == 11:
        return 6
    elif numero == 10:
        return 5
    elif numero == 7:
        return 4
    elif numero == 6:
        return 3
    elif numero == 5:
        return 2
    elif numero == 4:
        return 1

    return 0

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)
        
    def __repr__(self):
        return f"{self.numero} de {self.palo}"

c = Carta(7, "oro")
print(c)
```

**Mazo**: Representan las cartas espaÃ±olas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una funciÃ³n que permita barajar las cartas.
y otra funciÃ³n que permita repartir las cartas a los jugadores.

```python
import random

class Mazo:
    def __init__(self):

        numeros = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]
        palos = ["ðŸ—¡ï¸espada", "ðŸ¥’basto", "â˜€ï¸oro", "ðŸ·copa"]

        self.cartas= [Carta(numero, palo) for numero in numeros for palo in palos]
               
    def mezclar(self):
        random.shuffle(self.cartas)
    
    def repartir(self, cantidad):

        cartas_repartir = self.cartas[:cantidad]
        self.cartas = self.cartas [cantidad:]
        return cartas_repartir

m = Mazo()
m.mezclar()
print(m.repartir(3))
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
class Jugador:
    
    def __init__(self, nombre):
        self.nombre = nombre
        self.mano = []
        self.puntos = 0 
    def recibir_cartas(self, cartas):
        self.mano = cartas

    def jugar_carta(self):
        if self.mano:
            carta_jugada = self.mano.pop(0)
            return carta_jugada
        return None
    

m = Mazo()
m.mezclar()

jugador1 = Jugador("Juan")
jugador2 = Jugador("Pedro")

jugador1.recibir_cartas(m.repartir(3))
jugador2.recibir_cartas(m.repartir(3))

print(f"{jugador1.nombre}: {jugador1.mano}")
print(f"{jugador2.nombre}: {jugador2.mano}")
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
class Partida:

    def comparar_cartas(self, carta1, carta2):
        valor1 = carta1.valor
        valor2 = carta2.valor
        
        if valor1 > valor2:
            return 1 
        elif valor2 > valor1:
            return 2  
        else:
            return 0   

    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.mazo = Mazo()
        self.puntaje = {jugador1.nombre: 0, jugador2.nombre: 0}

    def jugar_ronda(self):

        self.mazo = Mazo()
        self.mazo.mezclar()
        cartas_jugador1 = self.mazo.repartir(3)
        cartas_jugador2 = self.mazo.repartir(3)
        self.jugador1.recibir_cartas(cartas_jugador1)
        self.jugador2.recibir_cartas(cartas_jugador2)


        print(f"\nðŸ“Puntaje Actual: {self.jugador1.nombre}: {self.puntaje[self.jugador1.nombre]}, {self.jugador2.nombre}: {self.puntaje[self.jugador2.nombre]}")
        print(f"{self.jugador1.nombre}: Cartas {[str(carta) for carta in self.jugador1.mano]}")
        print(f"{self.jugador2.nombre}: Cartas {[str(carta) for carta in self.jugador2.mano]}")

        puntos_jugador1 = 0
        puntos_jugador2 = 0


        for i in range(3):
            carta_jugador1 = self.jugador1.mano[i]
            carta_jugador2 = self.jugador2.mano[i]

            print(f"\n{self.jugador1.nombre} juega: {carta_jugador1} | {self.jugador2.nombre} juega: {carta_jugador2}")

            resultado = self.comparar_cartas(carta_jugador1, carta_jugador2)
            
            if resultado == 1:
                print(f"{self.jugador1.nombre} gana")
                puntos_jugador1 += 1
            elif resultado == 2:
                print(f"{self.jugador2.nombre} gana")
                puntos_jugador2 += 1
            else:
                print("Empate")

        if puntos_jugador1 > puntos_jugador2:
            print(f"\n{self.jugador1.nombre} gana la ronda")
            self.puntaje[self.jugador1.nombre] += 1
        elif puntos_jugador2 > puntos_jugador1:
            print(f"\n{self.jugador2.nombre} gana la ronda")
            self.puntaje[self.jugador2.nombre] += 1
        else:
            print("Ronda de empate")

    def jugar(self):
        while self.puntaje[self.jugador1.nombre] < 15 and self.puntaje[self.jugador2.nombre] < 15:
            self.mazo.mezclar()
            self.jugar_ronda() 
            
        if self.puntaje[self.jugador1.nombre] >= 15:
            print(f"ðŸ¥‡{self.jugador1.nombre} gana el juegoðŸ¥‡")
        else:
            print(f"ðŸ¥‡{self.jugador2.nombre} gana el juego!ðŸ¥‡")

p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
