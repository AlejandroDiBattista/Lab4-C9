## Ejercicio 1 

Escribe una función en Python que procese una lista de cadenas de texto y realice las siguientes acciones:

1. Tomar una lista de frases y convertir cada frase en una lista de palabras.
1. Luego, debe crear una nueva lista que contenga la cantidad de palabras en cada frase.
1. Tambien debe crear un lista que contenga la cantidad de caracteres en cada frase.
1. Finalmente, debe imprimir cada frase original junto con la cantidad de palabras que contiene.
Ejemplo de uso:
```python
    frases = [
        "Python es un lenguaje de programación",
        "Me gusta resolver problemas con código",
        "Las listas y los bucles son muy útiles"
    ]

    procesar_frases(frases)
```

ejemplo de salida:
```bash
    La frase: "Python es un..."
    tiene 6 palabras y 38 caracteres
```


````python
def convertir_palabras(frases):

#Creamos una lista para guardar variables
    lista_palabras=[]

#Recorremos cada frase en la lista    
    for frase in frases:
        palabras = frase.split()
        lista_palabras.append(palabras)

#Retornamos la lista de palabras    
    return lista_palabras

def contar_palabras(frases):
    
#creamos lista para almacenar la cantidad de palabras que tiene cada frase
    cantidad_palabras=[]
#Recorremos cada frase en la lista
    for frase in frases:
        palabras = frase.split() #Transformamos la frase en palabras
        cantidad_palabras.append(len(palabras))
    return cantidad_palabras # Devuelve la lista con la cantidad de palabras por frase

def frases_cantidad_palabras(frases):
    frases_cantidad=[]
    for frase in frases: 
        palabras = frase.split()
        cantidad_palabras= len(palabras)
        print (f'"{frase}" tiene {cantidad_palabras} palabras' )
    #Frases
frases = [
    "Python es un lenguaje de programación",
    "Me gusta resolver problemas con código",
    "Las listas y los bucles son muy útiles",
    "Javascript es un lenguaje de programación"
]    


convertir_frases = convertir_palabras(frases) 
print (convertir_frases)
cantidad_palabras = contar_palabras(frases)
print (cantidad_palabras)
frases_cantidad_palabras(frases)
````

## Ejercicio 2

Escribe una función en Python que encuentre los valores de `a`, `b`, y `c` para que la función cuadrática `f(x) = aX^2 + bX + c` pase exactamente por los siguientes puntos:

| x  | y  |
|---:|---:|
|  0 |  0 |
|  1 |  8 |
|  2 | 12 |
|  3 | 12 |
|  5 |  0 |

### Requisitos:
- La función debe explorar posibles valores de `a`, `b`, y `c` utilizando un método de prueba y error.
- Debe devolver los valores que hagan que la diferencia entre la función `f(x)` y los valores medidos `y` sea exactamente cero para cada punto.

> **Pista**: Los valores de `a`, `b`, y `c` son números pequeños.

````python
def funcion_cuadratica(a, b, c, x):
    return a * x**2 + b * x + c

puntos = [(0, 0), (1, 8), (2, 12), (3, 12), (5, 0)]

def encontrar_coeficientes():
    for a in range(-10, 11):
        for b in range(-10, 11):
            for c in range(-10, 11):
                es_correcto = True
                for x, y in puntos:
                    if funcion_cuadratica(a, b, c, x) != y:
                        es_correcto = False
                        break
                if es_correcto:
                    return a, b, c

a, b, c = encontrar_coeficientes()

print(f"Los valores encontrados son: a = {a}, b = {b}, c = {c}")
````
