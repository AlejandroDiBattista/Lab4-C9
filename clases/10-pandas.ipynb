```python
sueldos = [1000, 2000, 1500, 2000, 3000]
nombres = ['Juan', 'Pedro', 'Maria', 'Ana', 'Luis']
edades  = [20, 30, 20, 30, 20]
provincia = ['T','S','T','S','S']
s = pd.Series(sueldos)
n = pd.Series(nombres)
e = pd.Series(edades)
d = pd.DataFrame({
    'nombre': n, 
    'sueldo': s, 
    'edad': e,
    'provincia': provincia})
d.set_index('nombre', inplace=True)

d.groupby('edad')['sueldo'].mean()
d

# Agrupar por provincia y calcular los promedios de sueldo y edad
promedios = d.groupby('provincia').agg({
    'sueldo': 'mean',
    'edad': 'mean',
})

# Mostrar el resultado
promedios
```
